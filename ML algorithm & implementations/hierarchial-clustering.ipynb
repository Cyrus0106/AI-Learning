# Hierarchial Clustering Implementation in Python

## Hierarchical Clustering
Hierarchical clustering is an unsupervised learning algorithm that builds a hierarchy of clusters. It starts with each data point as its own cluster and then merges or splits cluster based on distance measures, forming a tree-like structure called a dendogram. The hierarchy can be used to choose a suitable number of clusters by "cutting" the tree at a specific level.

```python
# import necessary libraries
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np

# sample data (e.g points in 2D space)
X = np.array(
    [[1, 2], [6, 7], [2, 1], [5, 6], [3, 4], [4, 5], [1, 1], [2, 3], [4, 2], [3, 5]]
)

# perform hierarchial/agglorative clustering
Z = linkage(
    X, "ward"
)  # 'ward' is a distance metric it minimises the variance of the clusters

# plot the dendrogram
plt.figure(figsize=(8, 4))
dendrogram(Z)
plt.title("Dendogram for Hierarchial Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
