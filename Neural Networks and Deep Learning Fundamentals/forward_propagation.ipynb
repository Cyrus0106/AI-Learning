### Forward Proparagation and Activation Functions

Forward propagation is the process in a neural network where input data is passed through each layer, applying weights, biases, and activation functions, to produce an output prediction. It is the first step in training, used to calculate the predicted values before computing the loss and performing backpropagation.
**Steps in Forward Propagation:**
1. **Input Layer:**  
    Receive the input features.

2. **Weighted Sum:**  
    Multiply inputs by weights and add biases for each neuron.

3. **Activation Function:**  
    Apply an activation function (e.g., **ReLU**, **sigmoid**) to the weighted sum to introduce non-linearity.

4. **Hidden Layers:**  
    Repeat the weighted sum and activation process for each hidden layer.

5. **Output Layer:**  
    Compute the final output using the last set of weights, biases, and activation function.

6. **Prediction:**  
    The output layer produces the network's prediction.

7. **Including the One Before It:**  
    Each step builds upon the previous, ensuring information flows from input to output through all transformations.

    #### Common Activation Functions

    - **Sigmoid:**  
        Maps input values to a range between 0 and 1. Useful for binary classification.

    - **Tanh (Hyperbolic Tangent):**  
        Maps input values to a range between -1 and 1. Often used in hidden layers.

    - **ReLU (Rectified Linear Unit):**  
        Outputs zero for negative inputs and the input itself for positive values. Widely used due to simplicity and effectiveness.

    - **Leaky ReLU:**  
        Similar to ReLU but allows a small, non-zero gradient when the input is negative.

    - **Softmax:**  
        Converts a vector of values into probabilities that sum to 1. Commonly used in the output layer for multi-class classification.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define activation functions

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0,z)


def softmax(z):
    exp_z = np.exp(z - np.max(z))
    return exp_z / exp_z.sum(axis=0, keepdims=True)


# forward pass functions
def forward_pass(X, weights, biases, activation_activation):
    z = np.dot(weights, X) + biases
    a = activation_activation(z)
    return a


# example inputs
X = np.array([[0.5], [0.8]])
weights = np.array([[0.2, 0.4], [0.6, 0.1]])
biases = np.array([[0.1], [0.2]])

# perform forward pass with different activation functions
activations = {
    "Sigmoid" : sigmoid,
    "Tanh": tanh,
    "ReLU": relu,
    "Softmax": softmax
}

for name, func in activations.items():
    output = forward_pass(X,weights,biases,func)
    print(f"{name} Activation Output:\n{output}\n")

# define range of inputs
z = np.linspace(-10,10,100)

# plot activation functions
plt.figure(figsize=(12,8))
plt.plot(z,sigmoid(z),label="Sigmoid")
plt.plot(z,tanh(z),label="Tanh")
plt.plot(z,relu(z),label="ReLU")
plt.plot(z,softmax(z),label="Softmax")
plt.title("Activation Functions")
plt.xlabel("Input (z)")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```
